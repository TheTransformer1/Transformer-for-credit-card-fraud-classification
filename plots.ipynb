```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, PowerTransformer
import numpy as np
```

```python
file_path = 'creditcard_2013.csv'
data = pd.read_csv(file_path)
```

```python
X = data
```

```python
plt.figure(figsize=(20, 15))
for i, column in enumerate(X.columns):
    plt.subplot(7, 5, i + 1)
    sns.boxplot(x=X[column])
    plt.title(f'Box Plot of {column}')
    plt.tight_layout()

plt.show()
```

```python
amount_pt = PowerTransformer(method='box-cox')
X['Amount'] = amount_pt.fit_transform(X[['Amount']] + 1e-9)

tot_seconds = X['Time']
X['sin_tot_seconds'] = np.sin(2 * np.pi * tot_seconds / (24 * 60 * 60))
X['cos_tot_seconds'] = np.cos(2 * np.pi * tot_seconds / (24 * 60 * 60))
```

```python
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
plt.figure(figsize=(20, 15))
for i, column in enumerate(X.columns):
    plt.subplot(7, 5, i + 1)
    sns.boxplot(x=X[column])
    plt.title(f'Box Plot of {column}')
    plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(20, 15))
for i, column in enumerate(X_scaled.columns):
    plt.subplot(7, 5, i + 1)
    sns.boxplot(x=X_scaled[column])
    plt.title(f'Box Plot of {column} (Scaled)')
    plt.tight_layout()

plt.show()
```
